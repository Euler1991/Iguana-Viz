```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
drink = pd.read_csv('drinks_230130.csv',
                    index_col=False)
```

```python
drink.info()
```

```python
drink_colors = {'Agua':'#1ca3ec',
                'Coca-Cola':'#F40009',
                'Café':'#271300',
                'Jugo':'#ff8d00',
                'Cerveza':'#FAE96F',
                'Vino':'#63032e',
                'Nestea':'#BC481B',
                'Chocolate':'#bd9e8a',
                'Mezcal':'#8DC8B4',
                'Ron':'#DF9C40',
                'Licor 43':'#DF9C40'}
```

```python
drink_pivoted = drink.pivot(index="Día",
                            columns="Bebida",
                            values="Volumen").reset_index()
```

```python
drink_pivoted.fillna(0, inplace=True)
drink_pivoted
```

```python
drink_pivoted[drink_pivoted['Día']>15]['Café'].sum()
```

```python
y = drink_pivoted[drink_pivoted['Día']>15]
x = y['Día']
cols = ['Café', 'Chocolate', 'Coca-Cola', 'Nestea', 'Cerveza', 'Ron']
```

```python
y = drink_pivoted[drink_pivoted['Día']>22]
x = y['Día']
cols = ['Café', 'Coca-Cola', 'Nestea', 'Cerveza', 'Mezcal', 'Licor 43']

fig = plt.figure()
ax = plt.subplot(111)
fig.set_figheight(10)
fig.set_figwidth(15)

plt.bar(x,
        y['Agua'],
        color=drink_colors['Agua'],
        label='Agua')
b = y['Agua']

for c in cols:
    plt.bar(x,
            y[c],
            bottom=b,
            color=drink_colors[c],
            label=c)
    b += y[c]

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.set_title('¿Por qué bebes lo que bebes?',
             fontsize=24,
             pad=20)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(x, ['Lun 23', 'Mar 24', 'Mie 25','Jue 26', 'Vie 27', 'Sab 28', 'Dom 29'])
plt.yticks([0,1000, 2000, 3000, 4000, 4500], ['0L', '1L', '2L', '3L', '4L', '4.5L'])

ax.text(24, 3000,
        'Es que fui al \ncine a ver Aguatar',
        fontsize=16,
        horizontalalignment='center')

ax.annotate('',
            xy=(23, 1950),
            arrowprops=dict(arrowstyle='->'),
            xytext=(23.4, 3000))

ax.text(28, 4000,
        '¿Por qué regresé con un \nvaso nuevo (con chela) \na casa?',
        fontsize=16,
        horizontalalignment='center')

ax.annotate('',
            xy=(26.4, 3000),
            arrowprops=dict(arrowstyle='->'),
            xytext=(27.7, 4000))

ax.text(27.6, 2500,
        'Qué rica amaneció el agua de garrafón!!!',
        fontsize=16,
        horizontalalignment='left')

ax.annotate('',
            xy=(27.4, 1600),
            arrowprops=dict(arrowstyle='->'),
            xytext=(27.6, 2500))

plt.savefig('pblqb_230129.png', bbox_inches='tight')
plt.show()
```
